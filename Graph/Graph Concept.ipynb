```python
df=pd.DataFrame(np.random.randn(1000),columns=['data'])
```

```python
#using plot to plot the data in graph  in pandas dataframe.
#using figsize parameter to set the length and width of represented graph
df.plot(figsize=(20,10))
```

```python
df=pd.read_csv("https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv")
```

```python

df.plot()
```

```python
#we use kind argument to select the type of graph
df.plot(kind='bar')
```

```python
df.iloc[5:11].plot(kind='bar')
```

```python
#ploting bar graph for particular row using iloc

df.iloc[[5]].plot(kind='bar')
```

```python
#ploting histogranm chart
'''
in hist graph X
    x -axis represents range
    y -axis represdents frequency
'''
#it represents data in form of bins
df.plot(kind='hist')
```

```python
#ploting histogranm chart for particular column
df['sepal_length'].plot(kind='hist')

# to rotate the graph we use orientation parameter.
#df['sepal_length'].plot(kind='hist',orientation='horizontal')
```

```python
#using direct hist() function to plot instead of "kind" parameter in plot
df.hist(figsize=(10,10))
```

```python
#color parameter- to set color of graph
#alpha parameter- to set density or density of color

df.hist(figsize=(10,10),color='black',alpha=.20)
```

```python
#area plot 
#same like line plot
df.plot(kind='area',figsize=(20,10),alpha=.15)
```

```python
df
```

```python
#scatter plot
#we use scatter plot to find the relation ship between two variables
#in this graph the relationship between 'sepal_length' and 'petal_length' is strong
df.plot(kind='scatter',x='sepal_length',y='petal_length')
```

```python
#in this graph the graph is more scattered i.e 'sepal_length' and 'petal_width' relationshiop is not strong
df.plot(kind='scatter',x='sepal_length',y='petal_width')
```

```python
#scatterplot
#Here we are using c(color) parameter to represent another variable based on range of that column the color is represented
#s-paramter--increase the size of bubble

df.plot(kind='scatter',x='sepal_length',y='petal_length',c='sepal_width',s =100)
```

```python
#scatterplot
#Here we are using c(color) parameter to represent another variable based on range of that column the color is represented
#s-paramter--or we can represent size based on particular variable i.e here we are using 'petal_width' for representing the size of bubble

df.plot(kind='scatter',x='sepal_length',y='petal_length',c='sepal_width',s =df['petal_width']*100)
```

```python
#same as scatter plot but difference is style of representing i.e bubble into hexa shape
#gridsize parameter---is used to set the size of hexa shape.
df.plot(kind='hexbin',x='sepal_length',y='petal_length',gridsize=10)
```

**Graphs using Plotly**

```python
import cufflinks as cf

#Since it takes time to represent the graph we use the below method to make it fast representartion
cf.go_offline()
```

```python
df.iplot()
```

```python
#scatter plot
#in plotly "kind" parameter wont give scatterplot itg goives scatter graph with lines
#so we use "mode" parameter to represent scatter plot in circles.
df.iplot(x='sepal_length',y='petal_length',kind='scatter',mode='markers',color='green',size=10)
```

```python
#bubble plot
df.iplot(kind='bubble',x='sepal_length',y='petal_length',size='sepal_width')
```

```python
#scatter_matrix

df.scatter_matrix()
```

                                         **3D plot in plotly**

```python
'''
in below grpah the tooltip values are
x-given value in parameter
y-given value in parameter
z-given value in parameter
and last value represents index number in df
'''
df.iplot(kind='scatter3d',x='sepal_length',y='petal_length',z='petal_width')
```

```python
#bubble3d graph 
#in bubble 3d graph you have to add parameter "size" along with the three axis parameter for sure to get the final output
#or it will throw error.

df.iplot(kind='bubble3d',x='sepal_length',y='petal_length',z='petal_width',size='petal_width')
```
