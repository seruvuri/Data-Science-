**#Seaborn topic nots**

1.can represent data by group using **hue(color)** and **style(shape)** parameter 

2.Can plot seperate graphs at a time based on column values using **col** parameter 

3.can represent graph based on catagory using **catplot** function

```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df
```

```python
#scatter plot using seaborn

sns.scatterplot(df.sepal_length,df.petal_length,color='red')
```

```python
import matplotlib.pyplot as plt
```

```python
#scatter plot using matplotlib
plt.plot(df.sepal_length,df.petal_length,'o')
plt.xlabel("sepal_length")
plt.ylabel("petal_length")
```

```python
#histogram plot using matplotlib
plt.hist(df['sepal_length'])
```

```python
#histogram using seaborn
#in seaborn we use "distplot" function yo plot histogram
sns.distplot(df['sepal_length'])
```

```python
df1=sns.load_dataset('tips')
```

```python
df1
```

```python
#hue parameter----- will group the data based on the variable passed.


sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=df1)
```

```python
#Style parameter ------will group the data based on the variable passed in different shape for each category.


sns.scatterplot(x='total_bill',y='tip',style='smoker',data=df1,size='size')
```

```python
#col paramter-----plotting seperate graphs based on column value

sns.relplot(x='total_bill',y='tip',col='time',data=df1)
```

```python
#catogary plot in seaborn

sns.catplot(x='day',y='total_bill',data=df1)
```
